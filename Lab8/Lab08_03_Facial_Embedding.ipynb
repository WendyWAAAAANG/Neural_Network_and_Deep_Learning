# Compare the distance between two faces

## The model was trained in a way that faces with a distance of 0.6 or less should be a match. But if you want to be more strict, you can look for a smaller face distance. For example, using a 0.55 cutoff would reduce false positive matches at the risk of more false negatives.

## Note: This isn't exactly the same as a "percent match". The scale isn't linear. But you can assume that images with a smaller distance are more similar to each other than ones with a larger distance.

## Under the hood, we use numpy.linalg.norm (Matrix or vector norm). The function is able to return one of eight different matrix norms, or one of an infinite number of vector norms, depending on the value of the ord parameter.

# NOW match faces for Thomas!

```python
image_to_test = face_recognition.load_image_file("fe/30a1cff9-a479-4af9-9456-57f484cf2dc9.jpg")

image_to_test_encoding = face_recognition.face_encodings(image_to_test)[0]

face_distance = face_recognition.face_distance([xiumin_face_encoding], image_to_test_encoding)

print(face_distance)
```

# Find multiple pictures for Thomas

```python
from IPython.display import Image, display
print("A distance of 0.6 or less means a matching face")
for i in range(1, 7, 1):
    image_file = "fe/0" + str(i) + ".jpg"
    image_to_test = face_recognition.load_image_file(image_file)
    image_to_test_encoding = face_recognition.face_encodings(image_to_test)[0]
    face_distance = face_recognition.face_distance([xiumin_face_encoding], image_to_test_encoding)
    if(face_distance <= 0.6):
        print("Found XIUMIN in", image_file, "- distance:", face_distance)
        #display(Image(filename=image_file, width=400))
    else:
        print("Cannot find XIUMIN in", image_file, "- distance:", face_distance)
```

# Find TSNE of faces

```python
import numpy as np
from sklearn.manifold import TSNE

known_encodings = [
    xiumin_face_encoding,
    karina_face_encoding
]
names = ['XIUMIN', 'Karina']
n_classes = 2

reduced = TSNE(n_components=2, random_state=0).fit_transform(known_encodings)

import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus']=False
%matplotlib inline

plt.figure(figsize=(5, 5))
for i in range(n_classes):
    X_label = reduced[i]
    plt.scatter(X_label[0], X_label[1], label = names[i])
plt.legend()
```

# Often instead of just checking if two faces match or not, it's helpful to see how similar more faces are.

```python

image_to_test = face_recognition.load_image_file("lzm_encode/07.jpg")

image_to_test_encoding = face_recognition.face_encodings(image_to_test)[0]

face_distances = face_recognition.face_distance(known_encodings, image_to_test_encoding)

for i, face_distance in enumerate(face_distances):
    print("Distance {:.2} from known image #{}".format(face_distance, i))
    print("- With a cutoff of 0.6, would the test match the known? {}".format(face_distance <= 0.6))
    print()
```

# You can also analyze and visualize the faces by using 3D PCA

```python
known_encodings = [
    xiumin_face_encoding,
    karina_face_encoding
]
names = ['XIUMIN', 'Karina']
n_classes = 2

for i in range(1, 7):
    image_file = "xiumin_encode/0" + str(i) + ".jpg"
    image_to_test = face_recognition.load_image_file(image_file)
    image_to_test_encoding = face_recognition.face_encodings(image_to_test)[0]
    known_encodings.append(image_to_test_encoding)
    print("Processing file:", image_file)
```

```python
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d

import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus']=False

X_xiumin = [known_encodings[0]]
X_karina = [known_encodings[1]]

for i in range(2, 7):
    X_xiumin.append(known_encodings[i])  
```

```python
pca = PCA(n_components=3).fit(known_encodings)
Xd_xiumin = pca.transform(X_xiumin)
Xd_karina = pca.transform(X_karina)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
plt.rcParams['legend.fontsize'] = 10   
ax.plot(Xd_xiumin[:,0], Xd_xiumin[:,1], Xd_xiumin[:,2],
        'o', markersize=8, color='blue', alpha=0.5, label='XIUMIN')
ax.plot(Xd_karina[:,0], Xd_karina[:,1], Xd_karina[:,2],
        'o', markersize=8, color='red', alpha=0.5, label='Karina')

plt.title('3D PCA Face Embedding')
ax.legend(loc='upper right')

plt.show()
```
